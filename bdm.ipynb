```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("grocery_data.csv")
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
for i in df.index:
    cat = df["Category"][i]
    if(cat=="Oil & Masala" or cat=="Food Grain"):
        df["Profit"][i]=df["Sales"][i]/20
```

```python
df.sort_values(by = "Order Date", inplace=True)
```

```python
df.to_csv("grocery_data.csv")
```
